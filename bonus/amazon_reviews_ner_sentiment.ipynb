```python
import spacy
import pandas as pd

# Load spaCy
nlp = spacy.load("en_core_web_sm")

# Sample Review
text = "I love the Apple AirPods. They're far better than my old Samsung earbuds."

# Named Entity Recognition
doc = nlp(text)
for ent in doc.ents:
    print(f"{ent.text} ({ent.label_})")

# Simple Rule-Based Sentiment
positive_words = ['love', 'great', 'excellent', 'amazing']
negative_words = ['bad', 'worst', 'hate', 'terrible']

sentiment = "Positive" if any(word in text.lower() for word in positive_words) else "Negative"
print(f"Sentiment: {sentiment}")
```
